```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('fifa21_train.csv')
data
data.describe().T
```

```python
dtypes= data.dtypes
list (dtypes)
```

```python
mean_volleys = data['Volleys'].mean()
data['Volleys'] = data['Volleys'].fillna(mean_volleys)
mean_curve = data['Curve'].mean()
data['Curve'] = data['Curve'].fillna(mean_curve)
mean_agility = data['Agility'].mean()
data['Agility'] = data['Agility'].fillna(mean_agility)
mean_balance = data['Balance'].mean()
data['Balance'] = data['Balance'].fillna(mean_balance)
mean_jumping = data['Jumping'].mean()
data['Jumping'] = data['Jumping'].fillna(mean_jumping)
mean_interceptions = data['Interceptions'].mean()
data['Interceptions'] = data['Interceptions'].fillna(mean_interceptions)
mean_positioning = data['Positioning'].mean()
data['Positioning'] = data['Positioning'].fillna(mean_positioning)
mean_vision = data['Vision'].mean()
data['Vision'] = data['Vision'].fillna(mean_vision)
mean_composure = data['Composure'].mean()
data['Composure'] = data['Composure'].fillna(mean_composure)
mean_sliding_tackle = data['Sliding Tackle'].mean()
data['Sliding Tackle'] = data['Sliding Tackle'].fillna(mean_sliding_tackle)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
X=data.drop('OVA', axis=1)     
y=data['OVA']
```

```python
data_num = X.select_dtypes(np.number)
data_cat = X.select_dtypes(object)
data_num.shape
```

```python
data_num.isna().sum()
```

```python
transformer = MinMaxScaler().fit(data_num)
data_normalized = transformer.transform(data_num)
print(data_normalized.shape)
data_num_scale = pd.DataFrame(data_normalized, columns=data_num.columns)
data_num_scale.head()
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
X_train, X_test, y_train, y_test= tts(data_num, y, test_size=0.2, random_state=42)
X_train
```

```python
y_train
```

```python
#import library
from sklearn.linear_model import LinearRegression as linreg

lm = linreg()
lm.fit(X_train, y_train)  
y_pred = lm.predict(X_test)   
```

```python
y_pred
```

```python
print ('train R2: {} -- test R2: {}'.format(lm.score(X_train, y_train),
                                            lm.score(X_test, y_test)))
```

```python
from sklearn.metrics import mean_squared_error as mse


train_mse=mse(lm.predict(X_train), y_train)
test_mse=mse(y_pred, y_test)

print ('train MSE: {} -- test MSE: {}'.format(train_mse, test_mse))
```

```python
def cleanfifa (data):
    data = data.drop_duplicates().reset_index(drop=True)
    columns_to_fill = ['Volleys', 'Curve', 'Agility', 'Balance', 'Jumping', 'Interceptions', 'Positioning', 'Vision', 'Composure', 'Sliding Tackle']
    data.fillna({col: data[col].mean() for col in columns_to_fill}, inplace=True)
    return data
data = pd.read_csv('fifa21_validate.csv')
    
        
cleaned_data = cleanfifa(data)
```
